```python
path=r'/Users/claudia/Documents/CoastSat-master/data/MogotesTodo2010a2020'
df=pd.read_csv(path+r'/transect_time_series_tidally_corrected.csv',header=0,names=['dates','t1','t2','t3','t4','t5'])

df.head()
```

```python
t=['t1','t2','t3','t4,','t5']
df2019=df[68:89]
dates2019=df2019['dates']
df2019.index = pd.to_datetime(dates2019)
for i in range (1,5):
    t[i]=df2019['t'+str(i)] - np.nanmedian(df2019['t'+str(i)])
    df2019['t_corregida'+str(i)]=t[i]
df2019.index
```

```python
fig = plt.figure(figsize=(8,4))
ax = plt.axes()
for i in range (1,5):
    ax.plot(t[i],label='t'+str(i))
ax.set(xlabel="tiempo", ylabel="nivel [m]")
ax.grid(color='grey',linestyle=':')
plt.legend()

#porque me plotea 4 y no 5
```
